```python
plt.rcParams['figure.figsize'] = [16, 12]

fig, ax = plt.subplots()
ax.imshow(im, cmap=plt.cm.gray)

for props in regs:
    y0, x0 = props.centroid
    orientation = props.orientation
    x1 = x0 + np.cos(orientation) * 0.5 * props.minor_axis_length
    y1 = y0 - np.sin(orientation) * 0.5 * props.minor_axis_length
    x2 = x0 - np.sin(orientation) * 0.5 * props.major_axis_length
    y2 = y0 - np.cos(orientation) * 0.5 * props.major_axis_length

    ax.plot((x0, x1), (y0, y1), '-r', linewidth=2.5)
    ax.plot((x0, x2), (y0, y2), '-r', linewidth=2.5)
    ax.plot(x0, y0, '.g', markersize=15)

    minr, minc, maxr, maxc = props.bbox
    bx = (minc, maxc, maxc, minc, minc)
    by = (minr, minr, maxr, maxr, minr)
    ax.plot(bx, by, '-b', linewidth=2.5)

ax.axis((0, 512, 512, 0))
plt.show()
```

```python
temp_Ntrajectory = 20000

Lmax = 0

x = np.full((N_frames, temp_Ntrajectory), np.nan, dtype=float)
y = np.full((N_frames, temp_Ntrajectory), np.nan, dtype=float)
theta = np.full((N_frames, temp_Ntrajectory), np.nan, dtype=float)
majorR = np.full((N_frames, temp_Ntrajectory), np.nan, dtype=float)
minorR = np.full((N_frames, temp_Ntrajectory), np.nan, dtype=float)
M = np.full((N_frames, temp_Ntrajectory, 4, 4), np.nan, dtype=float)
bounds = np.full((N_frames, temp_Ntrajectory, 2, 5), np.nan, dtype=float)

for t in range(N_frames):
    im = io.imread(os.path.join(path_mask, tif_names_mask[t]))
    regs = regionprops(im)
    Lmax = np.maximum(Lmax, im.max())
    if Lmax > x.shape[1]:
        x = np.append(x, np.full((N_frames, temp_Ntrajectory), np.nan, dtype=float), axis=1)
        y = np.append(y, np.full((N_frames, temp_Ntrajectory), np.nan, dtype=float), axis=1)
        theta = np.append(theta, np.full((N_frames, temp_Ntrajectory), np.nan, dtype=float), axis=1)
        majorR = np.append(majorR, np.full((N_frames, temp_Ntrajectory), np.nan, dtype=float), axis=1)
        minorR = np.append(minorR, np.full((N_frames, temp_Ntrajectory), np.nan, dtype=float), axis=1)
        M = np.append(M, np.full((N_frames, temp_Ntrajectory, 4, 4), np.nan, dtype=float), axis=1)
        bounds = np.append(bounds, np.full((N_frames, temp_Ntrajectory, 2, 5), np.nan, dtype=float), axis=1)
    for props in regs:
        L = props.label
        y[t, L-1], x[t, L-1] = props.centroid
        
        theta[t, L-1] = props.orientation
        minorR[t, L-1] = props.minor_axis_length
        majorR[t, L-1] = props.major_axis_length
        M[t, L-1] = props.moments_normalized
        
        minr, minc, maxr, maxc = props.bbox
        bounds[t, L-1] = np.array([[minc, maxc, maxc, minc, minc], 
                                   [minr, minr, maxr, maxr, minr]])

x = x[:,:Lmax]
y = y[:,:Lmax]
theta = theta[:,:Lmax]
majorR = majorR[:,:Lmax]
minorR = minorR[:,:Lmax]
M = M[:,:Lmax,:,:]
bounds = bounds[:,:Lmax,:,:]
```

```python
plt.rcParams['figure.figsize'] = [16, 12]

path_figsave = os.path.join(path_mask, 'traj/')
os.makedirs(path_figsave, exist_ok=True)

cm = plt.get_cmap("jet")
n_traj_plot = 10
colors = cm(np.linspace(0,1,n_traj_plot+1))

for t in range(N_frames):
    im = np.zeros((512,512,3), dtype=np.float32)
    im[...,:2] = np.transpose(io.imread(os.path.join(path_orig, tif_names_orig[t]))/65535., (1,2,0))
    
    fig, ax = plt.subplots()
    ax.imshow(im)
    
    ax.plot(bounds[t,:,0,:].T, bounds[t,:,1,:].T, 'b-')
    
    t_plots_cand = t + np.arange(-n_traj_plot, 1, 1, dtype=int)
    flg_plots = np.logical_and(t_plots_cand >= 0, t_plots_cand < N_frames-1)
    
    t_plots = t_plots_cand[flg_plots]
    color_plots = colors[flg_plots]
    
    for i, tp in enumerate(t_plots):
        
        ax.plot(x[tp:(tp+2),:], y[tp:(tp+2),:], '-', color=color_plots[i])
    
    ax.set_xlim(-1, 513)
    ax.set_ylim(-1, 513)
    
    fig.savefig(os.path.join(path_figsave, os.path.splitext(tif_names_orig[t])[0])+'.png', format='png')
    
    plt.clf()
    plt.close()

```
